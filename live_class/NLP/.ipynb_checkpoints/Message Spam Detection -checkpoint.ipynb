### 1. Reading dataset and make it in proper format

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
!wget https://raw.githubusercontent.com/mohitgupta-omg/Kaggle-SMS-Spam-Collection-Dataset-/master/spam.csv
```

```python
sms = pd.read_csv('spam.csv',encoding='latin-1')
```

```python
sms
```

```python
sms.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
sms.rename(columns={'v1':'label','v2':'message'},inplace=True)
```

```python
sms.head()
```

```python
sms.shape
```

```python
sms['label'].value_counts()
```

```python
row_count = sms.shape[0]
```

```python
print('Ham % {} spam % {}'.format(
    round(len(sms[sms['label'] == 'ham'])/row_count,2),
    round(len(sms[sms['label'] == 'spam'])/row_count,2)
))
```

```python
sms['length'] = sms.message.str.len()
```

### 2. Encode labels

```python
# Label coding 0 and 1
sms['label'].replace({'ham':0, 'spam': 1},inplace=True)
```

### 3. Covert all cases to lower

```python
# Coverting all to lower
sms['message'] = sms['message'].str.lower()
```

```python
sms.head()
```

### 4. Remove punctuatations

```python
# Replace mail
sms['message'] = sms['message'].str.replace(r'^.+@[^\.].*\.[a-z]{2,}$','emailaddress', regex=True)

#Replace URL's with webaddress
sms['message'] = sms['message'].str.replace(r'^http\://[a-zA-Z0-9\-\.]+\.[a-zA-Z]{2,3}(/\s*)?$','webaddress', regex=True)

# Replace Money Symploe
sms['message'] = sms['message'].str.replace(r'£|\$','dollers', regex=True)

# Replace Phone number 
sms['message'] = sms['message'].str.replace(r'^\(?[\d]{3}\)?[\d]{3}[\s-]?[\d]{4}$','phonenumber', regex=True)

# Replace Numbers
sms['message'] = sms['message'].str.replace(r'\d+(\.\d+)?','numbr', regex=True)
```

```python
# Remove punctuation 
sms['message'] = sms['message'].str.replace(r'[^\w\d\s]',' ',regex=True)

# Replace whitespace between terms with a single space
sms['message'] = sms['message'].str.replace(r'\s+',' ',regex=True)

# Remove leading and trailing whitespace
sms['message'] = sms['message'].str.replace(r'^\s+|\s+?$', '',regex=True)
```

```python
sms.head()
```

### 5. Remove stopwords

```python
import string
import nltk
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english')+['u','ü','ur','4','2','im','dont' ,'doin' 'ure'])
stop_words
```

```python
sms['message'] = sms['message'].apply(lambda x: ' '.join(
    term for term in x.split() if term not in stop_words
))
```

```python
sms['clean_length'] = sms['message'].str.len()
```

```python
sms.head()
```

```python
print('Original Length',sms.length.sum())
print('Clean Length',sms.clean_length.sum())
```

### 6. Check stats for messages
### 7. Covert all text into vectors
### 8. import classifire 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
tf_vec = TfidfVectorizer()
naive = MultinomialNB()

feature = tf_vec.fit_transform(sms['message'])

X = feature
y = sms['label']
```

### 9. Train and Test
### 10. Check the accuracy/confusion matrix

```python
X_train,X_test, y_train,y_test = train_test_split(X,y,random_state=42,shuffle=True)

naive.fit(X_train,y_train)

y_pred = naive.predict(X_test)

print('Final Score: ', accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
# plot confusion heatmap
conf_mat = confusion_matrix(y_test,y_pred)

ax = plt.subplot()

sns.heatmap(conf_mat,annot=True,ax=ax,linewidths=5,linecolor='r',center=0)

ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(['ham','spam'])
ax.yaxis.set_ticklabels(['ham','spam'])
plt.show()
```
