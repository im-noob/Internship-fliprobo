```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn import preprocessing
```

```python
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv')
```

```python
df
```

```python
df = df[['petal.length','sepal.length']]
```

```python
df.head(3)
```

```python
model = KMeans(n_clusters=3)
```

```python
y_model = model.fit_predict(df)
```

```python
y_model
```

```python
model.labels_
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c=model.labels_,cmap='viridis')
```

```python
pd.value_counts(y_model)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df_iris = pd.DataFrame(iris.data)
```

```python
for i in range(2,10):
    clustring = AgglomerativeClustering(n_clusters=i)

    clustring_m = clustring.fit(df_iris)

    clustring_m

    clustring_m.labels_

    plt.scatter(df_iris.iloc[:,0],df_iris.iloc[:,1],c=clustring_m.labels_)
    plt.show()
```
