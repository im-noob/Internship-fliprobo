```python
for column in df.columns:
    sns.boxplot(y=df[column])
    plt.figure()
```

# Spliting feature and target column

```python
X = df.iloc[:,:-1]
X
```

# Checking for collinearity with VIF(Varance Inflaction Factor)

```python
df_info = pd.DataFrame()
df_info['VIF'] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
df_info['Column'] = X.columns
df_info.sort_values('VIF',ascending=False)
```

```python
# Remoivng Multicoliniery 

X = df.iloc[:,:-1]


X['diff_density_pH'] = X['density'] - X['pH'] - df['alcohol']
X.drop(['density','pH','alcohol'],axis=1,inplace=True)

df_info = pd.DataFrame()
df_info['VIF'] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
df_info['Column'] = X.columns
df_info.sort_values('VIF',ascending=False)
```

# Checking for skewness

```python
df.skew().sort_values()
```

```python
X_transformed = PowerTransformer().fit_transform(X)
X = pd.DataFrame(X_transformed,columns=X.columns)
X
```

```python
np.abs(X.skew()).sort_values()
```

```python
X.describe()
```

# SMOTE

```python
y = df.iloc[:,-1]
y.value_counts()
```

```python
smote = SMOTE(random_state=2)
X,y = smote.fit_resample(X,y.ravel())
```

```python
pd.DataFrame(y).value_counts()
```

```python
print(X.shape)
print(y.shape)
```

# train test split

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

# Creating model( LogisticRegression )

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

# Model Evalution

```python
y_predict = lr.predict(X_test)
y_predict
```

```python
confusion_matrix(y_test,y_predict)
```

```python
lr.score(X_test,y_test)
```

## Grid search CV Parameter Tuning

```python
param_grid = {
    'penalty': ['l1','l2'],
    'C': [0.001,0.01,0.1,1,10,100,1000],
    'solver':['liblinear']
}
model_lr = GridSearchCV(LogisticRegression(),param_grid=param_grid)
```

```python
model_lr.fit(X_train,y_train)
```

```python
model_lr.best_score_
```

```python
model_lr.best_params_
```

## Cross validaiton

```python
cv = KFold(n_splits=10,random_state=42,shuffle=True)
scores = cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
print('score:',np.mean(scores))
print('std:',np.std(scores))
```
